# SEGUNDO PASO



Creo la query conjunta con las consultas anteriores

```python
# En la consulta que hago, unifico todo a traves del identificador unico comun, y en el where no expecifico a que tabla pertence cada columna porque no se repiten
# si tuviera por ejemplo cantidad de baños en dos tablas ya si especificaria por ejemplo a.cant_banyos, pero como no es el caso no referencio a que tabla pertenecen

Consultaglobal = """
               SELECT *
               FROM alojamiento AS A 
               LEFT JOIN precio AS B ON A.codigo_alojamiento = B.codigo_alojamiento
               LEFT JOIN puntuacion AS C ON A.codigo_alojamiento = C.codigo_alojamiento
               LEFT JOIN ubicacion AS D ON A.codigo_alojamiento = D.codigo_alojamiento
               WHERE puntos >= 4.5 
               AND puntos_agencia >= 4
               AND porcentaje_reserva = 25
               AND precio_alquiler BETWEEN 1500 AND 2000 
               AND dist_metro > 1.0
               AND dist_centro > 2.0
               AND superficie >= 80
               AND salon = 'SI'
               AND cant_banyos >= 2
               AND terraza = 'SI'
               AND airec_acond = 'SI'
               AND piscina = 'NO'        
"""
pd.read_sql(Consultaglobal, conn)
```

# TERCER PASO

¿Cuál es la suma total de metros cuadrados de los alojamientos que tienen piscina?

```python
Consulta8 = """
          SELECT SUM(superficie) AS Total_metros_piscina
          FROM alojamiento
          WHERE piscina = 'SI'
"""  
pd.read_sql(Consulta8, conn)
```

¿Cuál es el alojamiento con mayor superficie?

```python
Consulta9 = """
          SELECT codigo_alojamiento, nombre, superficie 
          FROM alojamiento
          WHERE superficie = (
           SELECT MAX(superficie)
           FROM alojamiento
           )       
"""  
pd.read_sql(Consulta9, conn)
```

¿Cuál es el alojamiento con menor superficie?


```python
Consulta10 = """
          SELECT codigo_alojamiento, nombre, superficie 
          FROM alojamiento
          WHERE superficie = (
           SELECT MIN(superficie)
           FROM alojamiento
           )
           
"""  
pd.read_sql(Consulta10, conn)
```

¿Cuántos alojamientos tienen salón?

```python
Consulta11 = """
           SELECT COUNT(codigo_alojamiento) AS total_alojamientos_con_salon
           FROM alojamiento
           WHERE salon = 'SI'
"""
pd.read_sql(Consulta11, conn)
```

¿Cuántos alojamientos tienen terraza?


```python
Consulta12 = """
           SELECT COUNT(codigo_alojamiento) AS total_alojamientos_con_terraza
           FROM alojamiento
           WHERE terraza = 'SI'
"""
pd.read_sql(Consulta12, conn)
```

¿Cuántos alojamientos hay por cada país según nuestra tabla de UBICACIÓN?

```python
Consulta13 = """
           SELECT Pais, COUNT(codigo_alojamiento) as Total_Alojamientos
           FROM ubicacion
           GROUP BY pais
           ORDER BY pais ASC
"""  
pd.read_sql(Consulta13, conn)
```

¿Cuál es el promedio de precios de alquiler de todos los alojamientos según la
tabla PRECIO?

```python
Consulta14 = """
           SELECT AVG(precio_alquiler) AS media_precios_alquiler
           FROM precio 
"""  
pd.read_sql(Consulta14, conn)
```

¿Cuáles son las 3 agencias que tienen la mayor puntuación de agencia (campo a
usar: PUNTOS_AGENCIA) según la tabla PUNTUACION?


```python
Consulta15 = """
           SELECT agencia, puntos_agencia
           FROM puntuacion
           ORDER BY puntos_agencia DESC
           LIMIT 3
"""  
pd.read_sql(Consulta15, conn)
```

Muestra los alojamientos (código y nombre) que: Contengan la palabra ‘Piso’ en su
nombre o que tengan piscina y aire acondicionado, pero no pasen los 150 metros
de superficie o que tengan más de un baño

```python
# Solo es necesario las dos primeras columnas pero incluyo las demas para confirmar que esta bien planteado

Consulta16 = """
           SELECT codigo_alojamiento, nombre, piscina, airec_acond, superficie, cant_banyos
           FROM alojamiento
           WHERE (nombre LIKE '%piso%'
           OR (piscina = 'SI' AND airec_acond = 'SI'))
           AND superficie < 150 
           OR cant_banyos > 1
         
"""  
pd.read_sql(Consulta16, conn)
```

Muestra una consulta que muestre el código de alojamiento, el nombre, el precio y
una columna donde veamos el precio del alquiler con el símbolo del euro al final de
cada importe, esta nueva columna se debe llama PRECIO_MODIF


```python
Consulta17 = """
           SELECT A.codigo_alojamiento, A.nombre, B.precio_alquiler, CONCAT(precio_alquiler,'€') AS precio_modif
           FROM alojamiento AS A
           LEFT JOIN precio AS B ON A.codigo_alojamiento = B.codigo_alojamiento       
"""  
pd.read_sql(Consulta17, conn)
```

Muestra una columna sobre la tabla PRECIO que se llame CATEGORIA_PRECIO y
que indique la palabra ‘Bajo’ si el precio está entre 780 y 1000, ‘Medio’ si el precio
está entre 1000 y 1700 y ‘Alto’ si es mayor a 1700

```python
Consulta18A = """
           SELECT codigo_alojamiento, reserva, porcentaje_reserva, precio_alquiler,
           CASE 
            WHEN precio_alquiler BETWEEN 780 AND 1000 THEN 'Bajo'
            WHEN precio_alquiler BETWEEN 1000 AND 1700 THEN 'Medio'
            WHEN precio_alquiler > 1700 THEN 'Alto'
            ELSE 'Sin categoria'
            END AS Categoria_precio 
           FROM precio     
"""  
pd.read_sql(Consulta18A, conn)

```
